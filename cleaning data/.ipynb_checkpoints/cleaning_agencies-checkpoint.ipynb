```python
import pandas as pd
```

```python
agencies = pd.read_csv('federal-firearm-licensees.csv')
```

```python
counties = pd.read_csv('uscitiesv.csv')
```

```python
counties = counties[['city', 'state_id', 'state_name', 'county_name']]
```

```python
merged_data = pd.merge(agencies, counties,  how='left', left_on=['city','state_id'], right_on = ['city','state_id'])
```

```python
merged_data
```

```python
mergedNAN = merged_data[merged_data.isna().any(axis=1)]
```

```python
mergedNAN
```

```python
merged_data=merged_data.dropna()
```

```python
merged_data.to_csv('firearm_agencies.csv')
```
