```python
import pandas as pd
```

```python
populationDF = pd.read_csv('2014population2.csv', encoding = "ISO-8859-1")
```

```python
populationDF = populationDF[~populationDF.county.str.contains('city')]
```

```python
populationDF
```

```python
states = pd.read_csv('states2014.csv')
```

```python
full_data = pd.merge(states, populationDF, how='left', left_on=['state'], right_on = ['county'])
```

```python
full_data
```

```python
full_data.to_csv('states2014.csv', index=False)
```
