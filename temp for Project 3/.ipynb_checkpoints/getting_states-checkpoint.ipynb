```python
import pandas as pd
```

```python
state_df = pd.read_csv('uscitiesv.csv')
```

```python
state_df = state_df[['state_id','state_name']]
```

```python
state_df['comb'] = state_df['state_id'] + state_df['state_name']
```

```python
df = pd.DataFrame(state_df.comb.unique())
```

```python
df.to_csv('states.csv')
```
